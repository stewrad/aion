```python
import numpy as np
import matplotlib.pyplot as plt

# Define constellation points
indices = np.array([1, 3, 7, 5])
const = np.exp(2j * np.pi * indices / 8.0)

# Dibit labels corresponding to each index
dibit_labels = ['00', '01', '10', '11']

# Plot the constellation
plt.figure(figsize=(6,6))
plt.scatter(const.real, const.imag, color='blue')

# Add labels
for i, point in enumerate(const):
    plt.text(point.real + 0.02, point.imag + 0.02, dibit_labels[i], fontsize=12, ha='center')

# Make plot nice
plt.axhline(0, color='gray', linestyle='--')
plt.axvline(0, color='gray', linestyle='--')
plt.xlabel('In-Phase')
plt.ylabel('Quadrature')
plt.title('Constellation with Dibit Labels')
plt.grid(True)
plt.axis('equal')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 8-PSK constellation (full)
N = 8
const = np.exp(2j * np.pi * np.arange(N) / N)

# Indices you want to label
indices = np.array([7, 6, 2, 0, 4, 5, 1, 3])

# Labels in 3-bit binary (octal form)
labels = [format(i, '03b') for i in indices]  # e.g., 7 -> '111'

# Plot all points
plt.figure(figsize=(6,6))
plt.scatter(const.real, const.imag, color='lightgray', label='Other points')

# Highlight and label selected points
for idx, lbl in zip(indices, labels):
    plt.scatter(const[idx].real, const[idx].imag, color='blue')
    plt.text(const[idx].real + 0.05, const[idx].imag + 0.05, lbl, fontsize=12, ha='center', color='red')

# Axes and grid
plt.axhline(0, color='gray', linestyle='--')
plt.axvline(0, color='gray', linestyle='--')
plt.xlabel('In-Phase')
plt.ylabel('Quadrature')
plt.title('8-PSK Constellation with Selected Labels')
plt.grid(True)
plt.axis('equal')

plt.show()
```
